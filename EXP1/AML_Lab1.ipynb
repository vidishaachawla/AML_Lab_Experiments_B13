```python
# 1. NumPy: 1D → 2D → 3D reshaping
import numpy as np

arr_1d = np.arange(1, 13)
arr_2d = arr_1d.reshape(3, 4)
arr_3d = arr_1d.reshape(2, 2, 3)

print("1D Array:\n", arr_1d)
print("\n2D Array (3x4):\n", arr_2d)
print("\n3D Array (2x2x3):\n", arr_3d)
```

```python
# 2. Pandas: Series with custom indices and division
import pandas as pd

s1 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
s2 = pd.Series([2, 4, 5], index=['a', 'b', 'c'])

result = s1 / s2
print(result)
```

```python
# 3. Matplotlib: Red line graph
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, 'r-')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Red Line Graph")
plt.show()
```

```python
# 4. SciPy: Determinant, Eigenvalues, Eigenvectors
from scipy import linalg
import numpy as np

matrix = np.array([[2, 1],
                   [1, 3]])

det = linalg.det(matrix)
eigenvalues, eigenvectors = linalg.eig(matrix)

print("Determinant:", det)
print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
# 5. Statistics: Mean, Median, Mode, Variance, Standard Deviation
import numpy as np
from statistics import mode

data = [10, 20, 20, 30, 40]

mean = np.mean(data)
median = np.median(data)
mode_val = mode(data)
variance = np.var(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode_val)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```
